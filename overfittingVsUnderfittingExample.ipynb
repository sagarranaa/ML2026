```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
```

```python
#create dataset
X=np.array([1,2,3,4,5,6,7]).reshape(-1,1)
y=([2,3,45,6,7,8,9])
```

```python
# Linear Model
model=LinearRegression()
model.fit(X,y)
pred = model.predict(X)
```

```python
# Plot it
plt.scatter(X,y)
plt.plot(X,pred,color='red')
plt.show()
```

```python
# overfitting Example
# now create a complex model
poly_model = make_pipeline(
    PolynomialFeatures(degree=6),
    LinearRegression()
)

poly_model.fit(X,y)

pred2 = poly_model.predict(X)
```

```python
# Plot overfitting example
plt.scatter(X,y)
plt.plot(X,pred2,color='green')
plt.show()
```

```python
#### why overfitting happens
# Too complex model
# Too many features
# small dataset
# noise in data
```

```python
# Multiple Linear Regression
# formula y=b0+b1x
# multiple features
# y= b0+b1x1+b2x2+b3x3
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data = {
    "Experience":[1,2,3,4,5,6,7,8],
    "Education":[12,13,14,15,16,16,17,18],
    "Salary":[30000,35000,40000,50000,60000,65000,70000,80000]
}

df = pd.DataFrame(data)

df
```

```python
# Dataset meaning
# Experience +Education+salary
# Defining features and target
X=df[['Experience','Education']]
y=df['Salary']
```

```python
# Split dataset
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
#Train Model
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
model.predict([[9,18]])
```

```python
model.coef_
```

```python
# Model Evalution
from sklearn.metrics import r2_score

pred = model.predict(X_test)

r2_score(y_test,pred)
```
